```python
   #LINEPLOT

# lets me a plotting for specific country and one or more specific year 

countrydata = {
    'Year': [1991, 1992, 1993, 1994, 1995, 1991, 1992, 1993, 1994, 1995],
    'Country': ['France', 'France', 'France', 'France', 'France', 
                'India', 'India', 'India', 'India', 'India'],
    'Unemployment Rate': [8.7, 8.6, 8.9, 9.0, 8.8, 7.4, 7.5, 7.6, 7.7, 7.8]
}

dff=pd.DataFrame(countrydata)  

def unemployment_plot(country_nm, years):
    country_data = dff[(dff['Country']==country_nm) & (dff['Year'].isin(years))]    
    
    if country_data.empty:
        print(f"Given country: {country_nm} or years: {years} not found or incorrect.")
    else:
        plt.figure(figsize=(12,7))        
        plt.plot(country_data['Year'], country_data['Unemployment Rate'], marker='o', linestyle='-', color='y' , label=f'{country_nm}')
        plt.legend()
        plt.title("unempployment rate BTW country and their years", color="g")
        plt.xlabel(" YEARS ",color="g")
        plt.ylabel(" UNEMPLOYMENT RATE (%)" , color="g")
        plt.xticks(rotation=80)
        plt.grid()
        plt.show()
    
unemployment_plot("France",[1991,1992,1993])
```
